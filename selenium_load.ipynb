```python
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
from webdriver_manager.chrome import ChromeDriverManager
from selenium import webdriver


options = Options()
options.add_experimental_option("detach",True)

driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()),
                          options=options)
driver.get("https://www.yelp.com/biz/plant-food-wine-venice?start=1")

```

```python
reviews = driver.find_elements(By.CLASS_NAME,'review__09f24__oHr9V')

for review in reviews:
    star = review.find_element(By.XPATH,'//*[@id="reviews"]/section/div[2]/div/ul/li[1]/div/div[2]/div/div[1]/span/div/div[1]').text
    date = review.find_element(By.XPATH,'//*[@id="reviews"]/section/div[2]/div/ul/li[1]/div/div[2]/div/div[2]/span').text
    review = review.find_element(By.XPATH,'//*[@id="reviews"]/section/div[2]/div/ul/li[1]/div/div[4]').text
    print(star,date,review)
```

```python
reviews = driver.find_elements(By.CLASS_NAME,'review__09f24__oHr9V')

review_list=[]

for review in reviews:
    star = review.find_element(By.XPATH,'//*[@id="reviews"]/section/div[2]/div/ul/li[3]/div/div[2]/div/div[1]/span/div').get_attribute("aria-label")
    date = review.find_element(By.XPATH,'//*[@id="reviews"]/section/div[2]').text
    review = review.find_element(By.XPATH,'//*[@id="reviews"]/section/div[2]').text
    print(star,date,review)

    rev_item = {
       'star': star,
       'date': date,
       'review':review
    }

    review_list.append(rev_item)

df = pd.DataFrame(review_list)
print(df)
```
